```python
from sklearn import svm
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
import pandas as pd
import os
```

```python
# input filePath
# generate svm data in gen_data file
def svm_data_gen(filePath, genName):
#     data_path = '../gen_data/dynamic_feature_2-25.data'
    data_path = '../gen_data/' + genName
    with open(data_path, 'wt') as f_data: # 可读可写，没有则创建，有则覆盖
        flst = os.listdir(filePath)
        for item in flst:
            path = filePath + item
            ev = pd.read_csv(path)
            ev = ev.values
            ev = np.reshape(ev, (-1,))[1:]
            
            lst = ','.join(str(i) for i in ev) # 转str
            # debug output
#             print(ev)
#             print(lst)
#             print("==================")
            
    
#             debug
#             print(np.shape(ev), ev)
#             print(item[17:20])
#             item[17:20] 测试者的label
            lst = lst + "," + item[17:20] + "\n" #添加标签
            f_data.write(lst)

# ref:
# 字符串转数组
# str = '1,2,3'
# arr = str.split(',')

# 数组转字符串
# arr = ['a','b']
# str = ','.join(arr)

# arr = [1,2,3]
# str = ','.join(str(i) for i in b)
```

```python
# input filePath
# generate mlp data in gen_data file
def mlp_data_gen(filePath, genName):
#     data_path = '../gen_data/dynamic_feature_2-26.data'
    data_path = '../gen_data/' + genName
    with open(data_path, 'wt') as f_data: # 可读可写，没有则创建，有则覆盖
        flst = os.listdir(filePath)
        for item in flst:
            path = filePath + item
            ev = pd.read_csv(path)
            ev = ev.values
            ev = np.reshape(ev, (-1,))[1:]
            
            lst = ','.join(str(i) for i in ev) # 转str
            # debug output
#             print(ev)
#             print(lst)
#             print("==================")
            
            
            s1=item.split('_')[2]
            s2=s1.split('-')[0][1:]
            lst = lst + "," + s2 + "\n" #添加标签
            f_data.write(lst)

# ref:
# 字符串转数组
# str = '1,2,3'
# arr = str.split(',')

# 数组转字符串
# arr = ['a','b']
# str = ','.join(arr)

# arr = [1,2,3]
# str = ','.join(str(i) for i in b)

name = "dynamic_feature-2-26" # 更换不同数据只需要修改name的参数即可，其它变量无需改动
data_name = name + ".data"
filePath = '../raw_data/' + name + '/'
mlp_data_gen(filePath, data_name)
```
